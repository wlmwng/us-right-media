# 01-inca-prep-urls
**Purpose**: split outlets' URLs from Media Cloud by year.
The collection is parallelized by running a separate script for each year.

```python
import pandas as pd
import os
```

```python
# dir_tw = os.path.join('..', '..', 'data', '02-intermediate', '02-twitter')
dir_mc = os.path.join('..', '..', 'data', '02-intermediate', '03-mediacloud')
dir_inca = os.path.join('..', '..', 'data', '02-intermediate', '04-inca-prep')
```

```python
# tw = pd.read_pickle(os.path.join(dir_tw, 'politicians_tweeted_urls.pkl'))
mc = pd.read_pickle(os.path.join(dir_mc, 'mediacloud_urls.pkl'))
```

```python
mc.columns
```

```python
mc["publish_year"] = mc["publish_date"].dt.to_period("Y")
dfs_by_year = mc.groupby(["publish_year"])
```

```python
# Q: why is kibana showing this URL as published at December 31st 2015, 23:12:13.000?
# A: it is because Kibana by default displays based on the browser. See Management -> Kibana (Advanced Settings) -> Timezone for date formatting
#    After setting it to UTC, it display as expected as 2016.
mc.loc[(mc['outlet']=='Breitbart') & (mc['url_id']=='410967446')]
```

```python
dfs = []
total = 0
for period, df in dfs_by_year:
    year = period.year
    if year in range(2016,2021):
        df = df.reset_index(drop=True)
        dfs.append(df)
        total += len(df)
        print(f"{period.year}: {len(df)} URLs")
        df.to_pickle(os.path.join(dir_inca, f"mediacloud_urls_{year}.pkl"))
        
print(f"total: {total} URLs")
```
